```python
expenses = []
```

```python
def add_expense(exp_id, date, category, description, amount):
    record = {
        "ID": exp_id,
        "Date": date,
        "Category": category,
        "Description": description,
        "Amount": amount
    }
    expenses.append(record)
```

```python
add_expense(1, "01-01-2025", "Food", "Lunch", 120)
add_expense(2, "02-01-2025", "Travel", "Bus Fare", 40)
add_expense(3, "03-01-2025", "Books", "Notebook", 150)
add_expense(4, "04-01-2025", "Food", "Snacks", 60)
add_expense(5, "05-01-2025", "Bills", "Mobile Recharge", 299)
add_expense(6, "06-01-2025", "Travel", "Auto", 80)
add_expense(7, "07-01-2025", "Food", "Dinner", 180)
add_expense(8, "08-01-2025", "Shopping", "Clothes", 900)
add_expense(9, "09-01-2025", "Education", "Online Course", 500)
add_expense(10, "10-01-2025", "Misc", "Stationery", 90)
```

```python
for e in expenses:
    print(e)
```

```python
print("Number of rows:", len(expenses))
print("Number of columns:", len(expenses[0].keys()))
print("Columns:", list(expenses[0].keys()))
```

```python
with open("expense_data.txt", "w") as file:
    for e in expenses:
        file.write(str(e) + "\n")

print("Dataset saved successfully")

```
